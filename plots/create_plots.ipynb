```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

colors = ["royalblue", "olivedrab", "firebrick", "goldenrod", "lightseagreen", "mediumvioletred"]
```

# Figure 4: BER and MI estimates obtained by DeepSE-WF and other WF security estimators.

```python
results = pd.read_csv("values/wfes-wefde-mine.csv")
results.set_index(["Defense", "Traces"], inplace=True)

deepse_results = pd.read_csv("values/architecture_convergence.csv")
deepse_results.set_index(["Defense", "Model", "Traces"], drop=False, inplace=True)

fig, ax = plt.subplots(1, 1, figsize=(16, 9))

X = range(6)

width = 1.2
shift = 1.3
spacing = 5

defenses = ["NoDef", "wtfpad", "Front_T1", "Front_T2", "cs_buflo", "tamaraw"]

for idx, defense in enumerate(defenses):
    if defense == "NoDef":
        ax.bar(
            idx * spacing - shift,
            deepse_results.loc[defense, "df", 90]["BER"].item(),
            width=width,
            label="DeepSE-WF (DF)",
            color=colors[0],
        )
        ax.bar(
            idx * spacing,
            results.loc[defense, 90]["WFES"].item(),
            width=width,
            label="WFES",
            color=colors[1],
        )
        ax.bar(
            idx * spacing + shift,
            1 - deepse_results.loc[defense, "df", 90]["Accuracy"].item(),
            width=width,
            label="Tk-Tok Error",
            color=colors[3],
        )
    else:
        ax.bar(
            idx * spacing - shift,
            deepse_results.loc[defense, "df", 90]["BER"].item(),
            width=width,
            color=colors[0],
        )
        ax.bar(idx * spacing, results.loc[defense, 90]["WFES"].item(), width=width, color=colors[1])
        ax.bar(
            idx * spacing + shift,
            1 - deepse_results.loc[defense, "df", 90]["Accuracy"].item(),
            width=width,
            color=colors[3],
        )

defense_names = ["NoDef", "WTF-PAD", "Front_T1", "Front_T2", "CS-BuFLO", "Tamaraw"]
plt.xticks(ticks=np.arange(6) * spacing, labels=defense_names, fontsize=30)
plt.yticks(ticks=[0, 0.25, 0.5, 0.75, 1], fontsize=30)

plt.gca().spines[["top", "right"]].set_visible(False)
plt.grid(axis="y", color="lightgray", linestyle="solid")

plt.ylim(0, 1)

# plt.title('Defense Evalutation', fontsize=40)
plt.xlabel("Defense", fontsize=30)
plt.ylabel("Bayes Error Rate", fontsize=30)
plt.legend(fontsize=30, ncol=1)

plt.savefig("outputs/wfes-deepse-90.pdf")
plt.show()
```

# Figure 11: Placement of DeepSE-WF MI estimations within the Fano and Kovalevskij bounds for MI.

```python
def H(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)


def lower_bound(ber, w):
    return np.log2(w) - H(ber) - ber * np.log2(w - 1)


def upper_bound(ber, w):
    res = np.inf
    for k in range(1, w + 1):
        res = min(
            res, np.log2(w) - np.log2(k) - k * (k + 1) * np.log((k + 1) / k) * (ber - (k - 1) / k)
        )
    return res
```

```python
fig = plt.figure(figsize=(16, 9))

eps = 1e-6
x = np.linspace(0 + eps, 99 / 100, 1000)

plt.xlabel("Bayes Error Rate", fontsize=30)
plt.ylabel("Mutual Information", fontsize=30)

plt.plot(x, [upper_bound(xi, 100) for xi in x], linewidth=3, color=colors[2], label="Upper Bound")
plt.plot(x, [lower_bound(xi, 100) for xi in x], linewidth=3, color=colors[3], label="Lower Bound")

plt.gca().spines[["top", "right"]].set_visible(False)
plt.grid(color="lightgray", linestyle="solid")

plt.ylim(0, 7)

plt.yticks(fontsize=30)
plt.xticks(ticks=[0, 0.25, 0.5, 0.75, 1], labels=[0, 0.25, 0.5, 0.75, 1], fontsize=30)
plt.legend(fontsize=30)
plt.savefig("outputs/bounds-ber-mi.pdf")
plt.show()
```
